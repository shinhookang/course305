# Fourier Analysis



```python
import numpy as np 
import matplotlib.pyplot as plt 

```

```python
# ex16.1

T = 1.5 
w0 = 2*np.pi / T 
print(w0)

f = lambda t: 1.7 + np.cos(w0*t+1.0472)
```

```python
t= np.array([0, .15, .3, .45, .6, .75, .9, 1.05, 1.2, 1.35])
y = np.array([2.2, 1.595, 1.031, .722, .786, 1.2, 1.805, 2.369, 2.678, 2.614])
y2 = f(t)
print(y)
print(y2)
dt = 0.15
n = len(t)

sum_y = np.sum(y)
sum_yc = np.sum(y*np.cos(w0*t))
sum_ys = np.sum(y*np.sin(w0*t))
print('sum_y=',sum_y)
print('sum_yc=',sum_yc)
print('sum_ys=',sum_ys)

A0 = sum_y/n
print('A0=',A0)
A1 = sum_yc/n*2 
print('A1=',A1)
B1 = sum_ys/n*2 
print('B1=',B1)

# the least square fit becomes 
f2 = lambda t: A0 + A1*np.cos(w0*t) + B1*np.sin(w0*t) 

```

```python
tt = np.linspace(t[0],t[-1],101)
yy = f(tt)
zz = f2(tt)

plt.plot(tt,yy,c='b')
plt.plot(tt,zz,c='r',ls='--')
plt.xlabel('time')
plt.ylabel('y')
```

```python
# ex 16.3 and ex 16.4 
# Power Spectrum 
# https://docs.scipy.org/doc/scipy/tutorial/fft.html
from scipy.fft import fft
```

```python
#  compute DFT
n = 8
dt = 0.02
fs = 1./dt # sampling freq.
T = 0.16 # period
tspan = np.arange(0,n)/fs 
print(tspan)
freq1 = 12.5
freq2 = 18.75

func = lambda t: 5 + np.cos(2*np.pi*freq1*t) + np.sin(2*np.pi*freq2*t)
y = func(tspan)

print('sampling frequency, fs = {} [Hz]'.format(fs))
print('total sample length = ', n*dt)
print(n*dt == T)
print('Nyquist freq = f_max=',fs/2)
print('lowest detactable freq=', 1/(n*dt))
```

```python
tt = np.linspace(tspan[0],tspan[-1],101) 
yy = func(tt)
plt.plot(tt,yy,c='b')
plt.scatter(tspan,y,c='k',marker='s')
plt.xlabel('time')
plt.ylabel('y')

```

```python
Y = fft(y)/n
print(Y)
# Y[0] corresponds to the mean,
# Y[1:n/2] contain the positive-frequency terms, 
# Y[n/2:] contain the negative-frequency terms
freq = np.arange(1,n)/T
print(freq)
```

```python
# compute and plot the power spectrum 
nyquist = fs/2 
n2 = int(n/2)
fP = freq[0:n2]
Pyy = np.abs(Y[1:n2+1])**2
print(fP)
fig = plt.figure()
plt.stem(fP, Pyy, basefmt='k-', linefmt='k-', markerfmt='ko')
plt.grid()
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power Spectrum')
print(n2)
```

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.fft import fft 

# 1. Evaluate f(t) at 64 points from 0 to 2pi
npoint = 64
f = lambda t: np.cos(10.*t) + np.sin(3.*t)
t = np.linspace(0,2*np.pi,npoint)
ybase = f(t)
```

```python
# 2. Add random values to the signal 
noise = np.random.normal(0., 0.25, npoint) # 0 mean, sd=0.25

y = ybase + noise
y
```

```python
# plot y
fig=plt.figure()
ax1 = fig.add_subplot(211)
fig.subplots_adjust(hspace=0.5)
ax1.plot(tspan,y,c='k',marker='o')
ax1.grid()
ax1.set_xlabel('f(t) vs. time (s)')
```

```python
# 3. DFT 
Y = fft(y)/npoint
Y2 = fft(y,norm="forward")
Y - Y2
```

```python
# plot the magnitude of the DFT (power spectrum)
fig=plt.figure()
ax1 = fig.add_subplot(211)
fig.subplots_adjust(hspace=0.5)
ax1.plot(tspan,y,c='k',marker='o')
ax1.grid()
ax1.set_xlabel('f(t) vs. time (s)')
T = t[-1]
Ypow = abs(Y[1:])**2
freq = np.arange(1,npoint)/T
ax2 = fig.add_subplot(212)
ax2.stem(freq,Ypow,basefmt='k-',linefmt='k-',markerfmt='ko')
ax2.set_xlabel('Power vs. frequency (Hz)')
ax2.grid()
```
