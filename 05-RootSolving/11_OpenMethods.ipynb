```python
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt 

```

```python
def NewtonIteration(f,dfdx,xi):

    delta_x = - f(xi)/dfdx(xi)
    xp = xi + delta_x

    return xp
```

```python
# ex 6.4
f = lambda x: x - np.exp(-x)
dfdx = lambda x: 1.0 + np.exp(-x)


print('iteration, xi, f(xi), dfdx(xi), xp')

iter = 0
xi = 0.
xp = NewtonIteration(f,dfdx,xi)
print(iter, xi, f(xi), dfdx(xi), xp)

iter = 1
xi = xp
xp = NewtonIteration(f,dfdx,xi)
print(iter, xi, f(xi), dfdx(xi), xp)

iter = 2
xi = xp
xp = NewtonIteration(f,dfdx,xi)
print(iter, xi, f(xi), dfdx(xi), xp)


```

```python
print('iteration, xi, f(xi), dfdx(xi), xp')
xp = 0.
for iter in range(10):
    xi = xp
    xp = NewtonIteration(f,dfdx,xi)
    print(iter, xi, f(xi), dfdx(xi), xp)
```

```python
print('iteration, xi, f(xi), dfdx(xi), xp, et')
xtrue =  0.5671432904097838
xp = 0.
for iter in range(10):
    xi = xp
    xp = NewtonIteration(f,dfdx,xi)
    et = (xtrue-xi)/xtrue * 100.
    print(iter, xi, f(xi), dfdx(xi), xp, et)
```

```python
# ex 6.5 slow convergence 
f = lambda x: x**10. - 1.0
dfdx = lambda x: 10. * x**9. 

print('iteration, xi, f(xi), dfdx(xi), xp, et')
# xtrue =  0.5671432904097838
xp = 0.5
for iter in range(40):
    xi = xp
    xp = NewtonIteration(f,dfdx,xi)
    # et = (xtrue-xi)/xtrue * 100.
    # print(iter, xi, f(xi), dfdx(xi), xp, et)
    print(iter, xi, f(xi), dfdx(xi), xp)

```

```python
def newtraph(f, fp, x0, es=1.e-7, maxit=30):
    for i in range(maxit):
        x1 = x0 - f(x0)/fp(x0)
        ea = np.abs((x1-x0)/x1)
        if (ea < es): break 
        x0 = x1 
    return x1, f(x1), ea, i+1
```

```python
# ex 
f = lambda x: x**2. - 9.
dfdx = lambda x: 2.*x 

print('iteration, xi, f(xi), dfdx(xi), xp, et')
# xtrue =  0.5671432904097838
xi = 5.

(xp, fxp, ea, n) = newtraph(f, dfdx, xi, es=1.e-7, maxit=30)
print(xp, fxp, ea, n)

```
