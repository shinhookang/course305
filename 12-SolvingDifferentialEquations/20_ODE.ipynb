# Ordinary Differential Equations



```python
import numpy as np 
import matplotlib.pyplot as plt 

```

```python
# ex22.1,

# exact solution
f= lambda t: 4./1.3 * (np.exp(0.8*t) - np.exp(-0.5*t)) + 2*np.exp(-0.5*t) 

# Euler approximation
df= lambda t,y: 4. * np.exp(0.8*t) - 0.5*y

dt = 1. # stepsize 
t0 = 0.
tmax = 4. 

EulerStep = lambda dt,tn,yn: yn + dt * df(tn,yn)

y0 = f(t0)
t1 = t0 + dt
y1 = EulerStep(dt,t0,y0)
print(y1)

t2 = t1 + dt
y2 = EulerStep(dt,t1,y1)
print(y2)

t3 = t2 + dt
y3 = EulerStep(dt,t2,y2)
print(y3)

t4 = t3 + dt
y4 = EulerStep(dt,t3,y3)
print(y4)

y_euler = np.array([y0,y1,y2,y3,y4])
x_euler = np.array([t0,t1,t2,t3,t4])

xx = np.linspace(t0, tmax, 101)
yy = f(xx) 

# plt.scatter(x_euler,y_euler,c='k',marker='o')
plt.plot(x_euler,y_euler,c='k',marker='o',ls='--',label='Euler f(t)')
plt.plot(xx,yy,c='k',ls='-',label='true f(t)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
# ex 22.2 Heun's method without iteration

print('value at t=0 = {}'.format(f(t0) ))
print('slope at t=0 = {}'.format(df(t0,y0) ))


Predictor = EulerStep
# pyn = Predictor(dt,tn,yn)
Corrector = lambda dt,tn,yn,pyn: yn + dt * (df(tn,yn) + df(tn+dt,pyn) )/2.

# t=1 
t1 = t0 + dt 
py1 = Predictor(dt,t0,y0)
y1 = Corrector(dt,t0,y0,py1)
print('... t=0 ')
print('predictor = {}'.format(py1 ))
print('corrector = {}'.format(y1 ))

# t=2
t2 = t1 + dt 
py2 = Predictor(dt,t1,y1)
y2 = Corrector(dt,t1,y1,py2)
print('... t=1 ')
print('predictor = {}'.format(py2 ))
print('corrector = {}'.format(y2 ))

# t=3
t3 = t2 + dt 
py3 = Predictor(dt,t2,y2)
y3 = Corrector(dt,t2,y2,py3)
print('... t=2 ')
print('predictor = {}'.format(py3 ))
print('corrector = {}'.format(y3 ))

# t=4
t4 = t3 + dt 
py4 = Predictor(dt,t3,y3)
y4 = Corrector(dt,t3,y3,py4)
print('... t=4 ')
print('predictor = {}'.format(py4 ))
print('corrector = {}'.format(y4 ))

```

```python
y_heun = np.array([y0,y1,y2,y3,y4])
x_heun = np.array([t0,t1,t2,t3,t4])

xx = np.linspace(t0, tmax, 101)
yy = f(xx) 

# plt.scatter(x_euler,y_euler,c='k',marker='o')
plt.plot(x_heun,y_heun,c='k',marker='o',ls='--',label='Heun f(t)')
plt.plot(xx,yy,c='k',ls='-',label='true f(t)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
